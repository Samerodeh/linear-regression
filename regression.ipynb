# Using a regression model, split your data into train and test data.

```python
X = df['age'].values
Y = df['children'].values
X = X.reshape(-1, 1)
x_train, x_test, = train_test_split(X, train_size=0.7, test_size=0.3, random_state=100)
y_train, y_test = train_test_split(Y, train_size=0.7, test_size=0.3, random_state=100) 
```

# Fit your training split to the regression model.

```python
fit_regression_model = LinearRegression()
fit_regression_model.fit(x_train, y_train)
fit_regression_model.fit(x_test, y_test) 
y_predict = fit_regression_model.predict(x_test)
```

# Show your regression model’s score.

```python
score_train = (f'{fit_regression_model.score(x_train, y_train)*100}%')
print(score_train)
score_test = (f'{fit_regression_model.score(x_test, y_test)*100}%')
print(score_test)
```

# Draw at least three conclusions from your regression model.

## conclusions1

```python
plt.scatter(x_train, y_train, color='blue')
plt.scatter(x_test, y_test, color='red')
plt.plot(x_test, y_predict, color='black')
plt.xlabel('age')
plt.ylabel('children')
plt.show()
```
